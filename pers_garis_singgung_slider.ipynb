```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
# fungsi yang akan diturunkan pada suatu nilai x 
# f(x) = x^2 - 5x +4 

#diskritisasi x dengan dx = 0.01
dx = 0.1
x  = np.arange(-50,50,dx)
y  = np.zeros(len(x))
yx = np.zeros(len(x))

def fungsi(x):
  y = x**3 - 5*x + 4
  return y

y[0] = fungsi(x[0])
for i in range(len(x)-1):
  y[i+1] = fungsi(x[i+1])
  yx[i]= (y[i+1] - y[i])/dx
```

```python
dic = {'x':x, 
      'f(x)':y,
      "f'(x)":yx}
dic = pd.DataFrame(dic)
dic
```

Mencari Persamaan garis singgung $f(x)$

$y = f'(x)*(x - x_0) + y_0$

```python
def singgung(x, y, yx):
  a = np.arange(-50,50)
  b = np.zeros(50)
  b = yx*(a-x) + y
  return a, b

import ipywidgets as widgets

@widgets.interact(i=(0,len(x)-2))
def plotSinggung(i):
    a, b = singgung(x[i], y[i], yx[i])
    fig, ax = plt.subplots(figsize = (8,8))
    ax.vlines(x=x[i], ymin=min(y), ymax=y[i], colors='tab:red', ls=':', lw=1)
    ax.hlines(y=y[i], xmin=min(x), xmax=x[i], colors='tab:red', ls=':', lw=1)
    ax.plot(x,y, label= 'grafik $f(x)= x^3-5x+4$')
    ax.plot(a,b, label= 'garis singgung di titik (%.3f, %.3f)'%(x[i], y[i]))
    ax.set(xlim=(-50,50), ylim=(-125000,125000))
    ax.grid(linewidth=0.5, linestyle=':')
    ax.legend(loc='upper left')
    plt.title('Persamaan Garis Singgung di Suatu Titik', pad=10, fontsize=15, color='blue', loc='center')
    plt.xlabel('x', fontsize=13)
    plt.ylabel('y', fontsize=13)
    return plt.show()
```
